```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

data = requests.get("https://en.wikipedia.org/wiki/World_population")
datasoup = BeautifulSoup(data.text, "html.parser")
```

```python
t = datasoup.find_all('table')
table = t[3]
# table
```

```python
h = table.find_all('th')
h = [i.text.strip() for i in h]
headings = h[:4]
# headings
```

```python
d = {}
for i in headings:
    d[i] = []
d_len = len(d)
# d
# d_len
```

```python
rw = table.find_all('td')
rw = [i.text.strip() for i in rw]
# rw
```

```python
cols = []
for i in range(d_len):
  x = rw[i::5]
  cols.append(x)
# cols
```

```python
final_dict = {}
y = list(d.keys())
for i in range(d_len):
  final_dict[y[i]] = cols[i]
final_dict
```

```python
df = pd.DataFrame(final_dict)
# df
```

```python
df.to_csv('world_population.csv', index = False)
with open('world_population.csv', 'r') as f:
  print(f.read())
```
